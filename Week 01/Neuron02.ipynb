```python
import math
import numpy as np

class Neuron:
    def __init__(self, x):
        self.weights=np.random.random(len(x))
        self.bias=np.random.random(1)
        self.inputs=x
        #self.function_list={"ReLU","sigmoid"}
        #self.function=globals()[function]
    
    def ReLU(self, x):
        if x<=0:
          return 0
        else:
          return x

    def sigmoid(self, x):
        return 1 / (1 + math.exp(-x))

    def activate(self,function=''):
        
        # Calculate the weighted sum of inputs and bias
        weighted_sum = np.dot(self.inputs, self.weights) + self.bias
        print(weighted_sum)
        
        # Call funtion using string within a class
        func = getattr(self, function)

        # Apply the activation function (in this case, sigmoid)
        activation = func(weighted_sum)
        
        return activation
```

```python
inputs = [0.5, 0.7, -2, 5, 7]
neuron = Neuron(inputs)
output = neuron.activate(function='ReLU')
print(output)
```
